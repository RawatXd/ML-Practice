```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("bill_survey.csv")
df.head()
```

```python
df.City.nunique()
```

```python
sns.histplot(df[df.City == "Mumbai"].Bill,kde=True)
```

```python
sns.histplot(df[df.City == "Surat"].Bill,kde=True)
```

```python
df.groupby("City")
```

```python
sample_mean = []

for city , subdf in df.groupby("City"):
    sample_mean.append(subdf.Bill.mean())

len(sample_mean)
```

```python
sns.histplot(sample_mean,kde=True)
```
