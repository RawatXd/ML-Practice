```python
def get_square(numbers):
    result = []
    for n in numbers:
        result.append(n*n)
    return result
```

```python
def get_cube(numbers):
    result = []
    for n in numbers:
        result.append(n*n*n)
    return result
```

```python
my_list = [1,2,3,4,5]
get_square(my_list)
```

```python
my_list2 = [2,3,4,5]
get_cube(my_list2)
```
