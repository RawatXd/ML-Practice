```python
class RemoteControl:
    def __init__(self):
        # You must define the list you are trying to loop through
        self.channels = ["HBO", "CNN", "ABC", "ESPN"]
        self.index = -1

    def __iter__(self): # Fixed spelling from __itr__
        return self

    def __next__(self):
        self.index += 1
        if self.index == len(self.channels):
            raise StopIteration
        return self.channels[self.index]

rc = RemoteControl()
for channel in rc:
    print(channel)
```

```python
def get_company_data():
    d=[]
    for i in range (3):
        with open(f'Stock_{i+1}.txt','r') as f:
            content = f.read()
            yield content
    return d
```

```python
for company_data in get_company_data():
    print(company_data)
```
