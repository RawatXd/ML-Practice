```python
prices = []

with open("pricing.txt","r") as f:
    for line in f:
        ticker , country , price ,data = line.split('|')
        prices.append({
            'ticker' : ticker,
            'country' : country,
            'price' : float(price),
            'data' : data.strip()
        })

prices[:2]
```
