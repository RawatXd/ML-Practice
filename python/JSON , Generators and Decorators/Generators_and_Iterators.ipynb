```python
prices = []

with open("pricing.txt","r") as f:
    for line in f:
        ticker , country , price ,data = line.split('|')
        prices.append({
            'ticker' : ticker,
            'country' : country,
            'price' : float(price),
            'data' : data.strip()
        })

prices[:2]
```

```python
class RemoteControl:
    def __init__(self):
        self.channels = ["NDTV","CNN","HBO"]
        self.index = -1

    def __next__(self):
        self.index +=1
        if self.index == len(self.channels):
            raise StopIteration
        return self.channels[self.index]

    def __itr__(self):
        return self
```

```python
rc = RemoteControl()
for channel in rc:
    print(channel)
```
